# Architecture overview

The model is mainly constructed around 3D convolutions. This type of layer not only takes care of the spatial distribution of pixel but also of their temporal evolution. Afterwards, classical fully connected layer manages the classification from the features extracted by the convolutions.

```python
dir_model = f'/n/groups/patel/JbProst/Heart/Scripts/CrossVal/Age/{3}ch/Fold{0}/'
weights_file = [w for w in os.listdir(dir_model) if w.endswith('.h5')][0]
model = Sequential()
model= load_model(dir_model + weights_file, {'RMSE' : RMSE, 'R2_': R2_})
model.summary()
```

Even if the architecture provides good results (see next parts), the main drawback is the number of parameters (187 millions). The saved model weights 2.25 GB, its use requires large CPU and GPu memory and its training is extremely time consuming. With a *NVIDIAÂ® V100 GPU*, around a 1.5 months are need to reach convergence (with 46k video samples).
This implied a limited range of architecture variation (like creating paires of convolutions layers instead of a single one).

#### Get R$^2$ score

Gathers all predictions scores into a single matrix

```python
cols= pd.MultiIndex.from_product([['3ch', '4ch'], ['Train', 'Test', 'Val']])
df_scores = pd.DataFrame(index=['Fold'+str(i) for i in range(10)],
                 columns=cols)
df_predictions = pd.DataFrame()

for channel in ['3ch','4ch']:
    for kfold in range(10):
        for part in ['train','test','val']:
            try:
                dir_csv = '/n/groups/patel/JbProst/Heart/Scripts/CrossVal/Age/{}/Fold{}/{}predictions_augmented.csv'.format(channel,
                                                                                                                  kfold,
                                                                                                                 part.lower())
                #load predictions
                df_tmp = pd.read_csv(dir_csv, index_col='eid')
                
                df_tmp['chambers'] = channel
                df_tmp['fold'] = kfold
                df_tmp['partition'] = part
                
                #compute the residual
                df_tmp['res'] = abs(df_tmp.Age_raw - df_tmp.Pred )
                #gather predictions
                df_predictions = pd.concat((df_predictions, df_tmp))

                #compute R2 score
                r2_tmp = r2_score(df_tmp.Age_raw,df_tmp.Pred )
                df_scores.loc['Fold'+str(kfold), (channel, part.capitalize())]= r2_tmp

            except FileNotFoundError:
                print('/{}/Fold{}/{}predictions_augmented.csv is MISSING'.format(channel, 
                                                                                 kfold,
                                                                                 part.lower()))
```

### Ensemble model

Gather the test predictions of each folds for each chamber view and create a simple ensemble model (averaging the predicted age from the 3 and 4 chambers view)

```python
df_all =pd.DataFrame()
for channel in [3,4]:
    directory = '/n/groups/patel/JbProst/Heart/Scripts/CrossVal/Age/{}ch/'.format(channel)
    df_ch = retrieve_predictions(directory)
    df_ch.rename(columns={'Age_raw': 'Age{}'.format(channel)}, inplace= True)
    df_ch.rename(columns={'Pred': '{}ch'.format(channel)}, inplace= True)
    df_ch.set_index(('eid'), inplace=True)
    df_all= pd.concat((df_all, df_ch), axis=1)

df_all.rename(columns={'Age{}'.format(3): 'Age'}, inplace= True)
df_all.drop(columns='Age4', inplace=True)
df_all.dropna(inplace=True)
```

Mean scores over the folds

```python
df_all['Mean'] = (df_all['3ch'] + df_all['4ch'])/2 #df_all[['4ch','3ch']].apply(lambda x: np.mean(x), axis= 1) 

print('R2 {}'.format(r2_score(df_all['Age'],df_all['Mean'])))
print('RMSE {}'.format(np.sqrt(mean_squared_error(df_all['Age'],df_all['Mean']))))
```
